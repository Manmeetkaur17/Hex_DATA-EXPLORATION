```python
#Pair Plot
sns.pairplot(df,hue="species", markers=['o','s','d'], palette="hls")
plt.show()
```

```python
sns.boxplot(x="species",y="sepal_length",data=df)
plt.figure(figsize=(68,6))
plt.show()
```

```python
sns.violinplot(x="species",y="sepal_length",data=df)
plt.figure(figsize=(68,6))
plt.show()
```

```python
sns.violinplot(x="species",y="petal_length",data=df)
plt.figure(figsize=(68,6))
plt.show()
```
